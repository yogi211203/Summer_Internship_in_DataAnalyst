# Tuple
They are used to store multiple items in a single variable.

A tuple is a collection which is ordered and unchangeable.
It is written with round brackets.

Types of Tuple Functions

*   min()
*   max()
*   count()
*   index()
*   tuple()


```python
fruits = ("apple", "banana", "cherry")              # Standard Method
print(fruits)
```

```python
ints = tuple([1,2,3,4,5,9,7,2,0])                   # Function Method

ints                                                #calling function
```

```python
fruits = ("apple", "banana", "cherry","banana")                  # Allows Duplicates
print(fruits)
```

```python
# Count() Method

nums = (1, 3, 7, 8, 7, 5, 4, 6, 8, 5, 6, 6, 6, 6, 6, 6)

nums.count(6)                     #it will count number of 6
```

```python
# Index() Method

nums = (1, 3, 7, 8, 7, 4, 6, 5, 8, 5)

nums.index(5)                               #it will show 1st index of the 5
```

```python
# Min() & Max() Method

nums = (1, 3, 7, 99, 7, 5, 4, 6, 8, 5)

# min(nums)
max(nums)
```

# Dictionary
Dictionaries are used to store data values in key:value pairs.

A dictionary is a collection which is ordered*, changeable and do not allow duplicates.


* get()
* items()
* key()
* values()
* pop()
* popitem()
* update()
* clear()


```python
car = {"brand": "Ford","model": "Mustang","year": 1964}

# # Get() Method
print(car.get("year"))          #gives value of the key (year)

# # items() Method
print(car.items())              #gives all keys and values as output

# # key() Method
print(car.keys())               #gives only keys as output

# # values() Method
print(car.values())             #gives only values as output
```

```python
# pop() Method
# car
car.pop("year")             #specific item will be deleted in pop() method
print(car)

# popitem() Method

car.popitem()               # last item will be deleted in popitem() method
print(car)
```

```python
# update() Method

car.update({"color": "White"})         #update() method is to add items to dictionary
print(car)
```

```python
# clear() Method

car.clear()                  #it will delete all items from the dictionary
print(car)
```

# Set
Sets are used to store multiple items in a single variable.

Sets are written with curly brackets.

* add()
* pop()
* remove()
* difference()
* intersection()
* union()
* update()

```python
fruits = {"apple", "banana", "cherry"}

# add()

fruits.add("orange")
print(fruits)

# pop()

fruits.pop()            #it will delete random element from set
print(fruits)

# remove()

fruits.remove("cherry")         #it will delete specific element
print(fruits)
```

```python
fruits = {"apple", "banana", "cherry"}
company = {"google", "microsoft", "apple"}

# difference()

print(company.difference(fruits))       # Output will be from LHS of dot set only
```

```python
# intersection()

print(company.intersection(fruits))      # Print common element from both the sets
```

```python
# union()

print(company.union(fruits))        # Combine both sets together to form new set
```

```python
# upate()

company.update(fruits)
```
