# For Loop

*   It is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).
*   Using Break & Contiue keyword



```python
fruits = ["apple", "banana", "cherry",'pineappple','mango']

for i in fruits:
  print(i)
```

```python
for x in fruits:
  print(x)
  if x == "banana":               #it will iterate till banana and it will break from there
    break
```

```python
for x in fruits:
  if x == "cherry":              #it will iterate till cherry and then it will continue from cherry it will skip cherry
    continue
  print(x)
```

# While Loop

*   With the while loop we can execute a set of statements as long as a condition is true.
*   Using Break & Contiue keyword

```python
num = 1

while num < 6:
  print(num)
  num += 1
```

```python
num = 1

while num < 6:
  print(num)
  if num == 3:            #it will break iteration at 3
    break
  num = num + 1
```

```python
num = 0

while num < 10:
  num += 1
  if num == 5:          #it will skip 5 and will continue
    continue
  print(num)
```

# Python Functions

*   A function is a block of code which only runs when it is called.

*   Advanced Functions - lambda, map, zip

```python
def intro():                     # Function Defining
  print("Function is called ")

intro()                           #function calling
```

```python
def my_ID(name):                    # Function with Arguments
  print("21IT172_" + name)

my_ID("Himmanshu")                  #Calling function and passing Arugment
```

```python
#function with multiple arguments


def my_ID(name,surname):                    # Function with Arguments
  print("21IT172_" + name,""+surname)

my_ID("Himmanshu","Vasoya")                  #Calling function and passing Arugment
```

```python
def add(a,b,c):
  return a+b+c

add(4, 5, 5)
```

```python
#Lambda Function

add = lambda a,b,c: print(a+b+c)
add(4, 5, 5)
```

```python
# Map Function

nums = [1, 2, 3, 4, 5]

def sq(n):
    return n*n

# sq(2)

square = map(sq, nums)              #it will map nums(LIST) with the sq() function

print(list(square))
```

```python
# Zip Function

name = ["Himanshu", "Madhav", "Yogi"]
ID = ["21IT172","21IT169","21IT129"]

student = zip(ID, name)               #it will combine two LIST with corresponding index

print(list(student))
```
