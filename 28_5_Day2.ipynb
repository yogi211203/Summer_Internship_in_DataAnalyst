
# List and it's Methods

Lists are used to store multiple items in a single variable.

*   Lists are used to store multiple items in a single variable.
*   It is a collection which is ordered, changeable and allows duplicate members.

**List Methods**


append(), sort(), reverse(), remove(), pop(), insert(), index(), extend(), count(), copy(), clear()







```python
fruits = ['apple', 'cherry', 'banana']

fruits.append("orange")         #add element at the end of the list
fruits.sort()                   #to sort the list
fruits.reverse()                #reverse the list
fruits.remove("banana")         #used to remove particular element
fruits.pop(1)                   #used to remove by providing index
fruits.insert(1, "watermelon")  #add element at particular index
fruits.index("orange")          #to know index of an element

others = ['pineapple', 'strawaberry', 'guava','cherry','pineapple']

fruits.extend(others)           #to add 2 list
fruits.count("cherry")          #counts how many times particular element is present
fruits.copy()                   #it will create copy of the list (duplicate list)
fruits.clear()                  #used to delete whole list
```

# Python Conditional Statements


*   If , elif, else statements

```python
b , a = 33 , 200

# print(b>a)

if b > a:
  print("Entered")
else:
  print('Did not entered')
```

```python
aman , vinod = 50 , 80

if vinod > aman:
  print("vinod scores greater than aman")
elif aman == vinod:
  print("Both scores equally")
```
